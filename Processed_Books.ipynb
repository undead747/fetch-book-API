## Import processed dataset

```python
import pandas as pd
```

```python
# some time dataset is not perfectly clean, that's okay because we just need bookId as input of get API fucntion
# so we just ignore redundant variable
```

```python
books = pd.read_csv('./dataset/goodreadsbooks/books.csv',error_bad_lines=False)
```

```python
books
```

## Get 20 % top rating count books

```python
vote_counts = books[books['ratings_count'].notnull()]['ratings_count'].astype('int')
```

```python
m = vote_counts.quantile(0.8)
m
```

```python
chosen_books = books[books['ratings_count'] >= m]
```

```python
chosen_books
```

## Export our dataset

```python
chosen_books.to_csv("./dataset/processed_dataset/processed_books.csv", index = False)
```
