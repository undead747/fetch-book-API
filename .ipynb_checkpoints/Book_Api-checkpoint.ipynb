#### Get Original Language

```python
#librarything_api[381]['commonknowledge']['fieldList']['field']
```

```python
original_language = []

for i in librarything_api:
    try: 
        check = 0
        common_knowledge = i['commonknowledge']['fieldList']['field']
        for j in common_knowledge:
            if j['@displayName'] == 'Original language':
                check = 1
                original_language.append(j['versionList']['version']['factList']['fact'])
        if check == 0: original_language.append('English')
    except: original_language.append('unknown')
        
```

```python
books['original_language'] = original_language
```

```python
#count = 0

#for i in original_language:
#    if i == 'unknown': print(count)
#    count += 1        
```

#### Get Original Title

```python
original_title = []

for i in librarything_api:
    try: 
        check = 0
        common_knowledge = i['commonknowledge']['fieldList']['field']
        for j in common_knowledge:
            if j['@displayName'] == 'Original title':
                check = 1
                original_title.append(j['versionList']['version']['factList']['fact'])
        if check == 0: original_title.append(i['title'])
    except: 
        try : original_title.append(i['title'])
        except: original_title.append('unknown')
```

```python
books['original_title'] = original_title 
```

#### Get character 

```python
characters = pd.DataFrame(columns = ['name'])
```

```python
count = 0
for i in librarything_api:
    try:
        common_knowledge = i['commonknowledge']['fieldList']['field']
        for j in common_knowledge:
            if j['@displayName'] == 'People/Characters':
                book_characters = j['versionList']['version']['factList']['fact']
                for character in book_characters:
                    if character != " ": characters = characters.append({'name': character}, ignore_index = True)
    except: pass
    count += 1
```

```python
characters
```

#### Remove duplicate Character name value :

```python
characters.drop_duplicates(subset = "name", inplace = True)
characters.sort_values("name", inplace = True)
characters = characters.reset_index(drop = True)
```

```python
characters
```

```python
indices = pd.Series(characters.index, index = characters['name'])
```

```python
count = 0
for i in librarything_api:
    try:
        common_knowledge = i['commonknowledge']['fieldList']['field']
        for j in common_knowledge:
            if j['@displayName'] == 'People/Characters':
                book_characters = j['versionList']['version']['factList']['fact']
                for character in book_characters:
                    if character != " ": character_of_book = character_of_book.append({'bookID': count,'characterID': indices[character]}, ignore_index = True)
    except: pass
    count += 1
```

```python
author_goodread_ids = pd.DataFrame(columns = ['name','author_goodread_ID'])
```

##### Create Authors table

```python
authors = pd.DataFrame(columns = ['name','gender','hometown','born_at','died_at','about','image','influences'])
```

```python
for i in authors_api:
    gender = i['gender']
    if gender == None: gender = 'unknown'
    hometown = i['hometown']
    if hometown == None: hometown = 'unknown'
    born_at = i['born_at']
    if born_at == None: born_at = 'unknown'
    died_at = i['died_at']
    if died_at == None: died_at = 'unknown'
    about = i['about']
    if about == None: about = 'unknown'
    influences = i['influences']
    if influences == None: influences = 'unknown'
    authors = authors.append({'name':i['name'] , 'gender':gender , 'hometown':hometown, 'born_at': born_at, 'died_at': died_at, 'about': about,'image': i['large_image_url'], 'influences': influences }, ignore_index = True)
```

```python
authors
```

##### Get Relationship Between Book And Author

###### Get Authors Of Book 

```python
indices = pd.Series(author_goodread_ids.index, index = author_goodread_ids['name'])
```

```python
indices
```

```python
authors_of_book = pd.DataFrame(columns = ['bookID','authorID','role'])
```

```python
count = 0
for i in books_api:
    try:
        for j in i['authors']['author']:
            role = j['role']
            if role == None: role = 'author'
            authors_of_book = authors_of_book.append({'bookID':count , 'authorID':indices[j['name']] , 'role':role}, ignore_index = True)
    except:
        j = i['authors']['author']
        role = j['role']
        if role == None: role = 'author'
        authors_of_book = authors_of_book.append({'bookID':count ,'authorID':indices[j['name']] , 'role':role}, ignore_index = True) 
    count+= 1 
```
